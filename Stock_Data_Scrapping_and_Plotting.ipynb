```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import warnings
```

```python
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
```

```python
response = requests.get(url)
netflix_data_webpage = response.text  # Store the HTML content as text
```

```python
soup = BeautifulSoup(netflix_data_webpage, 'html.parser')
print(soup.title.text)
```

```python
netflix_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])
netflix_data.head()
```

```python
for row in soup.find("tbody").find_all('tr'):
  col = row.find_all("td")
  date = col[0].text
  Open = col[1].text
  high = col[2].text
  low = col[3].text
  close = col[4].text
  adj_close = col[5].text
  volume = col[6].text
  netflix_data = pd.concat([netflix_data,pd.DataFrame({"Date":[date], "Open":[Open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj_close], "Volume":[volume]})], ignore_index=True)
```
