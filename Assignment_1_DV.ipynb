```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    "species": ["Adelie"] * 50 + ["Chinstrap"] * 50 + ["Gentoo"] * 50,
    "island": ["Torgersen", "Dream", "Biscoe"] * 50,
    "bill_length_mm": [39.1, 39.5, 40.3, 36.7, 39.3] * 30,
    "bill_depth_mm": [18.7, 17.4, 18.0, 19.3, 20.6] * 30,
    "flipper_length_mm": [181, 186, 195, 193, 190] * 30,
    "body_mass_g": [3750, 3800, 3250, 3450, 3650] * 30,
    "sex": ["Male", "Female"] * 75,
}

penguins = pd.DataFrame(data)


sns.set_theme(style="whitegrid")

sns.pairplot(
    penguins,
    diag_kind="kde",
    hue="species",
    kind="reg",
    palette="rainbow",
    height=3
)

plt.suptitle("Penguins Dataset Pairplot (Regression)", y=1.02)
plt.show()

```
